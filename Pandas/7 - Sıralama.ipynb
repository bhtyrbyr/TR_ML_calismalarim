#  Sıralama
## Veri yapılarında sıralama işlemlerini yapılışı
### Açıklama
- Sıralama için .sort_index() fonksiyonu kullanılır.
- Stun bazlı sıralama için `axis = 1` parametresi verilir.
- Ters sıralama için `ascending = False` parametresi verilir.

#### Series

```python
s = pd.Series(range(5), index=["e","d","a","b","c"])
s
```

```python
s.sort_index()
```

#### DataFrame

```python
df1 = pd.DataFrame(np.arange(12).reshape(3,4), index=["iki","bir","üç"],
                  columns = list("DABC"))
df1
```

```python
df1.sort_index()
```

```python
df1.sort_index(axis = 1, ascending = False)
```

## Değere göre sıralama

#### Series

```python
s = pd.Series([5,2,-3,12,65,1])
s
```

```python
s.sort_values(ascending = False)
```

#### DataFrame

```python
import math as mt
```

```python
def f(x):
    #print(x)
    y = list()
    for i in x:
        y.append(np.abs(mt.floor(i * 5)))
    return y
```

```python
df = pd.DataFrame(np.random.randn(4,5), index=[3,5,1,2], columns=list("BARZE"))
df
```

```python
df = df.apply(f)
df
```

```python
df.sort_index()
```

- Kendi denemem

```python
print(df.A.sort_values())
print(df.A.sort_values().index)
df.reindex(df.A.sort_values().index) 
```

- Uygulama

```python
df.sort_values(by=["A","Z","E"])
```
